```python
data_Normal = np.loadtxt("./datas/origin/fault_data_0HP_IF.txt")  
sig1 = data_Normal[1000:1500]
plt.figure(figsize=(10, 3))
plt.plot(sig1, color='blue')

data_Norma2 = np.loadtxt("./datas/origin/fault_data_0HP_OF.txt")  
sig2 = data_Norma2[820:1320]
plt.figure(figsize=(10, 3))
plt.plot(sig2, color='blue')

data_Norma3 = np.loadtxt("./datas/origin/fault_data_0HP_BF.txt")  
sig3 = data_Norma3[700:1200]
plt.figure(figsize=(10, 3))
plt.plot(sig3, color='blue')

data_Norma4 = np.loadtxt("./datas/origin/fault_data_0HP_IO.txt")  
sig4 = data_Norma4[800:1300]
plt.figure(figsize=(10, 3))
plt.plot(sig4, color='lightblue')

data_Norma5 = np.loadtxt("./datas/origin/fault_data_0HP_IB.txt")  
sig5 = data_Norma5[600:1100]
plt.figure(figsize=(10, 3))
plt.plot(sig5, color='lightblue')

data_Norma6 = np.loadtxt("./datas/origin/fault_data_0HP_OB.txt") 
sig6 = data_Norma6[700:1200]
plt.figure(figsize=(10, 3))
plt.plot(sig6, color='lightblue')

data_Norma7 = np.loadtxt("./datas/origin/fault_data_0HP_IOB.txt")  
sig7 = data_Norma7[700:1200]
plt.figure(figsize=(10, 3))
plt.plot(sig7, color='lightblue')
```

```python
#构建故障类别的语义向量
t =5
s1=[]
s2=[]
s3=[]
s4=[]
s5=[]
s6=[]
s7=[]
threshold1 = np.max(sig1)/t
threshold2 = np.max(sig2)/t
threshold3 = np.max(sig3)/t
threshold = 0
if threshold1 > threshold2:
    threshold = threshold1
    if threshold1 > threshold3:
        threshold = threshold1
    else:
        threshold = threshold3
else:
    threshold = threshold2
    if threshold2>threshold3:
        threshold = threshold2
    else:
        threshold = threshold3
for i in range(500):
    #内圈
    if sig1[i]>=threshold and  sig1[i]<2*threshold:
        s1.append(1)
    elif sig1[i]>=2*threshold and sig1[i]<3*threshold:
        s1.append(2)
    elif sig1[i]>=3*threshold and sig1[i]<4*threshold:
        s1.append(3)
    elif sig1[i]>=4*threshold and sig1[i]<=5*threshold:
        s1.append(4)
    else:
        s1.append(0) 
    #外圈

    if sig2[i]>=threshold and  sig2[i]<2*threshold:
        s2.append(1)
    elif sig2[i]>=2*threshold and sig2[i]<3*threshold:
        s2.append(2)
    elif sig2[i]>=3*threshold and sig2[i]<4*threshold:
        s2.append(3)
    elif sig2[i]>=4*threshold and sig2[i]<=5*threshold:
        s2.append(4)
    else:
        s2.append(0)   
     #滚子   
    if sig3[i]>=threshold and  sig3[i]<2*threshold:
        s3.append(1)
    elif sig3[i]>=2*threshold and sig3[i]<3*threshold:
        s3.append(2)
    elif sig3[i]>=3*threshold and sig3[i]<4*threshold:
        s3.append(3)
    elif sig3[i]>=4*threshold and sig3[i]<5*threshold:
        s3.append(4)
    else:
        s3.append(0)


#内外圈          
for i in range(500):
    if s1[i] == 4 or s2[i] == 4:
        s4.append(4)
    elif s1[i] == 3 or s2[i] == 3:
        s4.append(3)
    elif s1[i] == 2 or s2[i] == 2:
        s4.append(2)
    elif s1[i] == 1 or s2[i] == 1:
        s4.append(1)
    
    else:
        s4.append(0)


        
        
#内圈滚子       
for i in range(500):
    if s1[i] == 4 or s3[i] == 4:
        s5.append(4)
    elif s1[i] == 3 or s3[i] == 3:
        s5.append(3)
    elif s1[i] == 2 or s3[i] == 2:
        s5.append(2)
    elif s1[i] == 1 or s3[i] == 1:
        s5.append(1)
    else:
        s5.append(0)



        
#外圈滚子       
for i in range(500):
    if s2[i] == 4 or s3[i] == 4:
        s6.append(4)
    elif s2[i] == 3 or s3[i] == 3:
        s6.append(3)
    elif s2[i] == 2 or s3[i] == 2:
        s6.append(2)
    elif s2[i] == 1 or s3[i] == 1:
        s6.append(1)
    else :
        s6.append(0)


        
        
#内外圈滚子       
for i in range(500):
    if s2[i] == 4 or s3[i] == 4 or s1[i] == 4:
        s7.append(4)
    elif s2[i] == 3 or s3[i] == 3 or s1[i] == 3 :
        s7.append(3)
    elif s2[i] == 2 or s3[i] == 2 or s1[i] == 2 :
        s7.append(2)
    elif s2[i] == 1 or s3[i] == 1 or s1[i] == 1 :
        s7.append(1)
    else:
        s7.append(0)
```

```python
semantic_all = [s1,s2,s3,s4,s5,s6,s7]
semantic_train = []
semantic_test = []

for i in range(2000):
    semantic_train.append(s1)
for i in range(2000):
    semantic_train.append(s2)
for i in range(2000):
    semantic_train.append(s3)
semantic_train = np.vstack(semantic_train)

for i in range(2000):
    semantic_test.append(s4)
for i in range(2000):
    semantic_test.append(s5)
for i in range(2000):
    semantic_test.append(s6)
for i in range(2000):
    semantic_test.append(s7)
semantic_test = np.vstack(semantic_test)

# print(np.array(semantic_all).shape)
# print(np.array(semantic_train).shape)
# print(np.array(semantic_test).shape)

np.save('./datas/semantic_test/semantic_all', semantic_all)
np.save('./datas/semantic_test/semantic_train', semantic_train)
np.save('./datas/semantic_test/semantic_test', semantic_test)
```
